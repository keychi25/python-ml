# 前処理とスケール変換

## 3.3.1 さまざまな前処理

- StandardScaler：個々の特徴量の平均が０，分散が１になるように変換（全ての特徴量の大きさを揃える）．
- RoBustScaler：ここの特徴量が一定の範囲に収まるように変換（分散ではなく中央値と四分位数を用いる）．
- MinMaxScaler：データがちょうど0から１に収まるように変換．
- Normalizer：個々のデータポイントを，特徴量ベクトルｇユークリッド長１になるように変換．  
→ データポイントを半径１（より高次元なら超球面）に投射する．

```python
import mglearn

mglearn.plots.plot_scaling()
```

## 3.3.2 データ変換の適用


```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
cancer = load_breast_cancer() # 癌データを使用

X_train, X_test, y_train, y_test = train_test_split(
    cancer.data, cancer.target, stratify=cancer.target, random_state=1)

print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train_scaled = scaler.transform(X_train)

##　ここから
print("transformed shape: {}".format(X_train_scaled.shape))
print("per-feature minimum before scaling:\n {}".format(X_train.min(axis=0))) # 変換前の各特徴量の最小値
print("per-feature maximum before scaling:\n {}".format(X_train.max(axis=0))) # 変換前の各特徴量の最大値
print("per-feature minimum after scaling:\n {}".format(X_train_scaled.min(axis=0))) # 変換後の各特徴量の最小値
print("per-feature maximum after scaling:\n {}".format(X_train_scaled.max(axis=0)))  # 変換後の各特徴量の最大値
```

```python
# テストデータを変換
X_test_scaled = scaler.transform(X_test)

# 訓練データとテストデータに全く同じ変換を施すため，0から1に収まらないものもある．
print("per-feature minimum after scaling:\n{}".format(X_test_scaled.min(axis=0))) # 変換後の各特徴量の最小値
print("per-feature maximum after scaling:\n{}".format(X_test_scaled.max(axis=0))) # 変換後の各特徴量の最大値
```

## 3.3.3 訓練データとテストデータを同じように変換

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

X, _ = make_blobs(n_samples=50, centers=5, random_state=4, cluster_std=2)
X_train, X_test = train_test_split(X, random_state=5, test_size=.1)

fig, axes = plt.subplots(1, 3, figsize=(13, 4))

axes[0].scatter(X_train[:, 0], X_train[:, 1], c=mglearn.cm2(0), label="Training set", s=60)
axes[0].scatter(X_test[:, 0], X_test[:, 1], marker='^', c=mglearn.cm2(1), label="Test set", s=60)
axes[0].legend(loc='upper left')
axes[0].set_title("Original Data")

scaler = MinMaxScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

axes[1].scatter(X_train_scaled[:, 0], X_train_scaled[:, 1], c=mglearn.cm2(0), label="Training set", s=60)
axes[1].scatter(X_test_scaled[:, 0], X_test_scaled[:, 1], marker='^', c=mglearn.cm2(1), label="Test set", s=60)
axes[1].set_title("Scaled Data")

# ダメな例
# テストセットを訓練セットとは別にスケール変換
# 実際にはしてはいけない
test_scaler = MinMaxScaler()
test_scaler.fit(X_test)
X_test_scaled_badly = test_scaler.transform(X_test) 

axes[2].scatter(X_train_scaled[:, 0], X_train_scaled[:, 1], c=mglearn.cm2(0), label="training set", s=60)
axes[2].scatter(X_test_scaled_badly[:, 0], X_test_scaled_badly[:, 1], marker='^', c=mglearn.cm2(1), label="test set", s=60)
axes[2].set_title("Improperly Scaled Data")

for ax in axes:
    ax.set_xlabel("Feature 0")
    ax.set_ylabel("Feature 1")
fig.tight_layout()

plt.show()
```

## 3.3.4 教師あり学習における前処理の効果

```python
from sklearn.svm import SVC

X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, random_state=0)

svm = SVC(C=100)
svm.fit(X_train, y_train)
print("Test set accuracy: {:.2f}".format(svm.score(X_test, y_test)))
```

```python
# スケール変換の前処理を加える（MinMaxScaler）
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

svm.fit(X_train_scaled, y_train)

print("Scaled test set accuracy: {:.2f}".format(svm.score(X_test_scaled, y_test)))
```

```python
# スケール変換の前処理を加える（StandardScaler）
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

svm.fit(X_train_scaled, y_train)

print("SVM test accuracy: {:.2f}".format(svm.score(X_test_scaled, y_test)))
```
